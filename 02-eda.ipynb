```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds = pd.read_csv("cleaned_data.csv")
```

```python
companywisecount = ds.groupby("company")["company"].count()
plt.figure(figsize=(15, 5))
ax = companywisecount.plot(kind='bar', color='skyblue', title='Company-wise Count')
plt.xticks(rotation=45)
for i, v in enumerate(companywisecount):
    ax.text(i, v + 0.1, str(v), ha='center', va='bottom')
plt.xlabel('Company')
plt.ylabel('Count')
plt.show()
```
